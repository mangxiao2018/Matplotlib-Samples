# The Lifecycle of a Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {'Beijing': 109438.50,
        'Hebei': 103569.59,
        'Tianjin': 112214.71,
        'Shandong': 112591.43,
        'Henan': 100934.30,
        'Hunan': 103660.54,
        'Zhejiang': 137351.96,
        'Jiangsu': 123381.38,
        'Guangdong': 135841.99,
        'Guangxi': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
```

```python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

ax.plot(group_names,group_data,label='linear')
#设置一条水平红虚线来求平均
ax.axhline(group_mean, ls='--', color='r')                   
# plt.style.use('fivethirtyeight')
```
